```python
import pandas as pd
```

```python
df_it = pd.read_csv('proofreading_silvia_antonis - proof_reading_italian.csv',usecols=['index','sent_more_it','sent_less_it'],nrows=250).set_index('index')
```

```python
df_it
```

```python
df_en = pd.read_csv('data/crows_pairs_anonymized.csv',index_col='Unnamed: 0')
```

```python
df_bias = df_en.query("bias_type=='gender'").reset_index(drop=True)
```

```python
df_merged = df_bias.merge(df_it,left_index=True,right_index=True,how='right')
```

```python
df_merged.head()
```

```python
df_clean = df_merged.drop(['sent_more','sent_less'],axis=1).rename(columns={'sent_more_it':'sent_more','sent_less_it':'sent_less'})
```

```python
df_clean.head()
```

```python
df_clean.to_csv('cp_it.csv')
```
